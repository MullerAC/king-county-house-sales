- break year out from date column
- turn sqft_basement into int or float, turn '?' into 0, change sqft_basement into 1 or 0
- check for nan: turn to 0's
- make years_since_renovated column
- categoricals: bedrooms, bathrooms, floors, waterfront, view, condition, grade, yr_built, yr_renovated, zip

```python
df['yr_sold'] = df.date.map(lambda x: int(x.split('/')[-1]))
df
```

```python
df.sqft_basement.value_counts()
```

```python
df.replace({'sqft_basement': {'?': '0.0'}}, inplace=True)
df.sqft_basement = pd.to_numeric(df.sqft_basement)
```

```python
df.isna().sum()
```

```python
df.yr_renovated.value_counts()
```

```python
df.fillna(0.0, inplace=True)
```

```python
df.isna().sum()
```

```python
df['yr_since_renovation'] = np.where(df['yr_renovated']==0.0, df['yr_sold']-df['yr_built'], df['yr_sold']-df['yr_renovated'])
df['yr_since_built'] = df['yr_sold'] - df['yr_built']
```

```python
df
```

to remove:
- date
- view
- sqft_above
- sqft_living15
- sqft_lot15

```python
df.drop(['view', 'sqft_above', 'sqft_living15', 'sqft_lot15', 'date'], axis=1, inplace=True)
```

```python
df
```

```python
df['renovated'] = df.yr_renovated.map(lambda x: 1 if x>0 else 0)
df.head()
```

```python
df.to_csv('data/cleaned_data.csv', index=False)
```
